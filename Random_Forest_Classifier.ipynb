```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection  import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/heart (1).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X=df.drop(['target'],axis=1)
y=df['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
LR=LogisticRegression()
SVC=SVC()
DTC=DecisionTreeClassifier()
model1=LR.fit(X_train,y_train)
model2=SVC.fit(X_train,y_train)
model3=DTC.fit(X_train,y_train)
y_pred1=model1.predict(X_test)
y_pred2=model2.predict(X_test)
y_pred3=model3.predict(X_test)
print(accuracy_score(y_test,y_pred1))
print(accuracy_score(y_test,y_pred2))
print(accuracy_score(y_test,y_pred3))
```

```python
estimators=[('lr',model1),('svc',model2),('dtc',model3)]
for i in estimators:
  score=cross_val_score(i[1],X,y,cv=10,scoring='accuracy')
  print(i,score.mean())
```

RANDOM FOREST CLASSIFIER IMPLEMETATION

```python
from sklearn.ensemble import RandomForestClassifier
RDC=RandomForestClassifier()
```

```python
RDC.fit(X_train,y_train)
ypred5=RDC.predict(X_test)
accuracy_score(y_test,ypred5)
```

```python
score2=cross_val_score(RDC,X,y,cv=10,scoring='accuracy')
print(score2.mean())
```

```python
rdc2=RandomForestClassifier(n_estimators=200,
                            max_depth=10,
                            bootstrap= True,
                            oob_score= True,
                            n_jobs=-1,  verbose=1)
rdc2.fit(X_train,y_train)
ypred=rdc2.predict(X_test)
accuracy_score(y_test,ypred)
```

```python

params={'n_estimators':[100,200,300],
        'max_depth':[3,5,7,9],
        'bootstrap':[True,False],
        'oob_score':[True,False]
}
GDS=GridSearchCV(param_grid=params,estimator=rdc2,cv=10,n_jobs=-1,scoring='accuracy')
print(GDS)
```

```python
GDS.fit(X_train,y_train)
```

```python
print(GDS.best_score_)
print(GDS.best_params_)
print(GDS.best_estimator_)
print(GDS.refit_time_)
print(GDS.best_index_)
print(GDS.n_splits_)
print(GDS.scorer_)
```
