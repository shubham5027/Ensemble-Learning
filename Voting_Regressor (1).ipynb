```python
X=df.drop(['Extracurricular Activities','Performance Index'],axis=1)
```

```python
y=df['Performance Index']
```

```python
x_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
model1=LinearRegression()
model2=DecisionTreeRegressor()
model3=SVR()
```

```python
from sklearn.metrics import SCORERS
sklearn.metrics.get_scorer_names()
```

```python
estimators=[('lr',model1),('dtr',model2),('SVR',model3)]
for i in estimators:
  scores=cross_val_score(i[1],X,y,cv=5,scoring='r2')
  print(scores.mean())
```

```python
from sklearn.ensemble import VotingRegressor
for i in range(2,10):
  vt=VotingRegressor(estimators=estimators)
  score=cross_val_score(vt,X,y,cv=i,scoring='r2')
  print(i,score.mean())
```
